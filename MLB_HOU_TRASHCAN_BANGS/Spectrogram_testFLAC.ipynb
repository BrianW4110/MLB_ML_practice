```python
# opens one flac file and converts to spectrogram, next try mfcc
import matplotlib.pyplot as plt
import numpy as np
import soundfile as sf # for opening FLAC
import librosa # for MFCC
from scipy.signal import spectrogram

file1 = "flac/2017_06_29_oakmlb_houmlb_1_41.flac"
audio_data, sample_rate = sf.read(file1)

frequencies, times, Sxx = spectrogram(audio_data, sample_rate)

plt.pcolormesh(times, frequencies, 10 * np.log10(Sxx))
plt.colorbar(label = "dB")
plt.ylabel("Frequency (Hz)")
plt.xlabel("Time (s)")
plt.title("Spectrogram of 2017_06_29_oakmlb_houmlb_1_41")
plt.show()
```

```python
# open multiple FLAC files from a directory
import os
import matplotlib.pyplot as plt
import numpy as np
import soundfile as sf # for opening FLAC
import librosa # for MFCC
from scipy.signal import spectrogram

path_dir = "flac"

for filename in os.listdir(path_dir):
    file_path = os.path.join(path_dir, filename)
    audio_data, sample_rate = sf.read(file_path)

    frequencies, times, Sxx = spectrogram(audio_data, sample_rate)

    plt.pcolormesh(times, frequencies, 10 * np.log10(Sxx))
    plt.colorbar(label = "dB")
    plt.ylabel("Frequency (Hz)")
    plt.xlabel("Time (s)")
    plt.title((file_path.split("\\"))[1].split(".")[0])
    plt.show()
    
```
