```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.io import wavfile
from scipy.signal import spectrogram

file1 = "2017_06_29_oakmlb_houmlb_1_41.wav"
sample_rate, audio_data = wavfile.read(file1)

window_size = 1024
hop_size = 512 # amount of overlap between segments

frequencies, times, Sxx = spectrogram(audio_data, sample_rate)

plt.pcolormesh(times, frequencies, 10 * np.log10(Sxx))
plt.colorbar(label = "dB")
plt.ylabel("Frequency (Hz)")
plt.xlabel("Time (s)")
plt.title("Spectrogram of 2017_06_29_oakmlb_houmlb_1_41")
plt.show()
```
